# ML Lab – Experiment 2  
## Data Preprocessing and Visualization  
### Dataset: Cervical Cancer Risk Factors


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/content/risk_factors_cervical_cancer 2.csv')
df.head()
```

```python
df.info()
```

```python
df = df.replace('?', np.nan)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
df['Age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Biopsy', y='Age', data=df)
plt.title('Age vs Cervical Cancer Biopsy Result')
plt.show()
```

```python
sns.countplot(x='Smokes', data=df)
plt.title('Smoking Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap After Preprocessing')
plt.show()
```
