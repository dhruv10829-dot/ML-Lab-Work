# ML Lab – Experiment 3  
## Descriptive Statistical Analysis  
### Dataset: Cervical Cancer Risk Factors


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/content/risk_factors_cervical_cancer 2.csv')
df.head()
```

```python
df = df.replace('?', np.nan)
df = df.apply(pd.to_numeric, errors='coerce')
df = df.fillna(df.mean())
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```

```python
df.mean(), df.median(), df.mode().iloc[0]
```

```python
df.std(), df.var()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Observation:
- Descriptive statistics summarize the distribution and variability of cervical cancer risk factors.
- Age, number of pregnancies, and STD-related features show notable variation.
- Correlation analysis highlights relationships among medical and behavioral risk factors.
